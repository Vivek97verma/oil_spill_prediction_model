## selecting the independent and dependent variable

```python
x=df3
y=df['target']
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

### split the data into training and testing data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.8,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## create a function to evaluate the model and test the model score

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
def eval_mod(y_test,y_pred):
    cm=confusion_matrix(y_test,y_pred)
    print(cm)
    print('Accuracy score',accuracy_score(y_test,y_pred))
    print('classification report',classification_report(y_test,y_pred))

def mscore(model):
    Training_score= model.score(x_test,y_test)
    Testing_score= model.score(x_train,y_train)
    res_df=pd.DataFrame({'Training score':Training_score,'Testing score':Testing_score},index=[model])
    return res_df
```

# Apply the Logistic Regresssion model

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression(max_iter=1000,solver='liblinear')    # max_iter is used for maximum iteration and liblinear is used for avoid warnings.
reg.fit(x_train,y_train)
```

```python
y_pred_log=reg.predict(x_test)
```

```python
eval_mod(y_pred_log,y_test)
```

```python
mscore(reg)
```

# apply KNeighbors classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
```

```python
y_pred_knn=knn.predict(x_test)
```

```python
eval_mod(y_pred_knn,y_test)
```

```python
mscore(knn)
```

# Applying Decision Tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion='gini',max_depth=12,min_samples_split=10)
dt.fit(x_train,y_train)
```

```python
y_pred_dt=dt.predict(x_test)
```

```python
eval_mod(y_pred_dt,y_test)
```

```python
mscore(dt)
```

# Apply Random Forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=15,criterion='gini',max_depth=15,min_samples_split=15)
rf.fit(x_train,y_train)
```

```python
y_pred_rf=rf.predict(x_test)
```

```python
eval_mod(y_pred_rf,y_test)
```

```python
mscore(rf)
```

# apply bagging and adaboost

# combine all result- To know the best model

```python
all_result=pd.concat([mscore(reg),mscore(knn),mscore(dt),mscore(rf),mscore(bagg),mscore(adboost)])
all_result
```
